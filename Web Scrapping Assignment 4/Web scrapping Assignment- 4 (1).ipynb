## 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year

```python
driver = webdriver.Chrome(r'C:\Users\sivap\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url = 'https://archive.ics.uci.edu/'
driver.get(url)
driver.maximize_window()
```

```python
#click on show all repositories

URL=driver.find_element(By.XPATH,'/html/body/table[2]/tbody/tr/td/span/b/a').get_attribute('href')
driver.get(URL)
```

```python
Dataset=[]
Datatype= []
Task=[]
Attribute=[]
No_of_instances=[]
Attribute_no=[]
Year=[]


for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a'):
    Dataset.append(i.text)
time.sleep(2)
for y in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p'):
    Datatype.append(y.text)
time.sleep(2)
for g in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p'):
    Task.append(g.text)
time.sleep(2)

for t in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p'):
    Attribute.append(t.text)
time.sleep(2)

for r in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p'):
    No_of_instances.append(r.text)
time.sleep(2)

for v in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p'):
    Attribute_no.append(v.text)
time.sleep(2)

for x in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p'):
    Year.append(x.text)
time.sleep(2)
```

```python
print(len(Dataset),len(Datatype),len(Task),len(Attribute),len(No_of_instances),len(Attribute_no),len(Year))

#header is included in some of the scrapped data which shall be remove while creating a dataframe
```

```python
pd.DataFrame({'Dataset':Dataset,
             "Data TypeS":Datatype[1:],
             'Default Task':Task[1:],
              "Attribute types":Attribute[1:],
              "Number of instances":No_of_instances[1:],
              "Number of attributes":Attribute_no[1:],
              "Year":Year[1:]})
```
