```python
def fetch_images(query:str):
    google_url ="https://www.google.co.in/search?q={q}&tbm=isch&ved=2ahUKEwjLgfycv5z5AhWYhNgFHaFsBxIQ2-cCegQIABAA&oq={q}&gs_lcp=CgNpbWcQDFAAWABgAGgAcAB4AIABAIgBAJIBAJgBAKoBC2d3cy13aXotaW1n&sclient=img&ei=E_viYousIZiJ4t4PodmdkAE&authuser=0&bih=714&biw=1536&hl=en"
    driver.get(google_url.format(q=query))
    #creating empty list
    thumbnails = []
    img_url = []
    
    for thumbnail in driver.find_elements(By.CSS_SELECTOR,'img.Q4LuWd'): 
        if len(thumbnails)>=25:
            break
        else:
            thumbnails.append(thumbnail)

    for img in thumbnails:
        img.click() #to get good resolution images
        time.sleep(3)
        
        image = driver.find_elements(By.XPATH,'//*[@id="Sva75c"]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div[3]/div/a/img')
        for s in image:
            if s.get_attribute('src') and 'https' in s.get_attribute('src'):
                img_url.append(s.get_attribute('src'))
        
    return img_url
       
```

# Q9 Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver= webdriver.Chrome(r'C:\Users\sivap\Downloads\chromedriver_win32\chromedriver.exe')
url = 'https://www.youtube.com/watch?v=CuU9q2VKOyc'
driver.get(url)
driver.maximize_window()
```

```python
previous_height= driver.execute_script("return 0, document.documentElement.scrollHeight")
previous_height
```

```python
#scrolls until the end of the page
while True:
    driver.execute_script('window.scrollTo(0, document.documentElement.scrollHeight);')
    time.sleep(4)
    new_height = driver.execute_script("return 0, document.documentElement.scrollHeight")
    
    if new_height == previous_height:
        break
    previous_height = new_height
    
```

```python
Comments = []
Upvotes = []
Time = []

#scrape comments
for comment in WebDriverWait(driver,60).until(EC.visibility_of_all_elements_located((By.XPATH,'//*[@id="content-text"]'))):
    if  comment.text is None:
        Comments.append('-')
    else:
        Comments.append(comment.text.replace('\n', ''))
    
#scrape comment upvotes
for upvote in driver.find_elements(By.XPATH,'//span[@id="vote-count-middle"]'):
        Upvotes.append(upvote.text)
        
#scrape time when comment was posted
for t in driver.find_elements(By.XPATH,'//*[@id="header-author"]/yt-formatted-string/a'):
        Time.append(t.text)
```

```python
hostel_urls = []
privatesprice = [] 
dormsprice =[]
name=[]
distance = []
while True:
    for i in driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[1]/div[4]/div/div/div/div/a[2]'):
        hostel_urls.append(i.get_attribute('href'))
    #price from dorms    
    try:
        for dp in driver.find_elements(By.XPATH,'//*[@id="__layout"]/div/div[1]/div[4]/div/div/div/div[4]/a[1]/div[2]|//*[@id="__layout"]/div/div[1]/div[4]/div/div/div/div[3]/a[1]/div[2]|//*[@id="__layout"]/div/div[1]/div[4]/div/div/div/div[4]/a[1]/div[2]'):
            dormsprice.append(dp.text.replace('Dorms From\n', ' ').strip())
    except NoSuchElementException:
        dormsprice.append('-')
    
    #name 
    
    try:
        for n in driver.find_elements(By.XPATH,'//h2[@class="title title-6"]'):
            name.append(n.text)
    except NoSuchElementException:
        name.append('-')
    
    #distance from city
    try:
        for s in driver.find_elements(By.XPATH,'//span[@class="description"]'):
            distance.append(s.text)
    except NoSuchElement:
        distance.append('-')

   #price from privates
    try:
        for p in driver.find_elements(By.XPATH,'//*[@id="__layout"]/div/div[1]/div[4]/div/div/div/div[3]/a[1]/div[1]|//*[@id="__layout"]/div/div[1]/div[4]/div/div/div/div[4]/a[1]/div[1]'):
            privatesprice.append(p.text.replace('Privates From\n', ' ').strip())
            
    except NoSuchElementException:
        privatesprice.append('-')
    
    time.sleep(5)
    
    try:
        driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        Next.click() #Next button
    except NoSuchElementException:
        break
    
    
   
       
```

```python
ratings = []
total_reviews = []
overall_reviews= []
facilities = []
description=[]


for x in hostel_urls:
    driver.get(x)
    time.sleep(2)
    
    #ratings
    try:
        r = driver.find_element(By.XPATH,'//div[@class="score orange big"]')
        ratings.append(r.text)
    except NoSuchElementException:
        ratings.append('-')
    
    #total reviews
    try:
        t = driver.find_element(By.XPATH,'//div[@class="reviews"]')
        total_reviews.append(t.text.split(' ')[0])
    
    except NoSuchElementException:
        total_reviews.append('-')
    
    #overall reviews
    try:
        o = driver.find_element(By.XPATH,'//div[@class="keyword"]')
        overall_reviews.append(o.text)
    except NoSuchElementException:
        overall_reviews.append('-')
    
    #description:
    
    #click on show more
    try: 
        driver.find_element(By.XPATH,'//div[@class="toggle-content"]')
    except NoSuchElementException:
        pass
    try:
        d = driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        description.append(d.text.replace('\n',' '))
    except NoSuchElementException:
        description.append('-')
    
    ##facilities
    
    
    try:
        f= []
        for facility in driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[1]/section/div[10]/div/ul/li/ul/li|/html/body/div[3]/div/div/div[1]/section/div[11]/div/ul/li/ul/li|/html/body/div[3]/div/div/div[1]/section/div[9]/div/ul/li/ul/li'):
            f.append(facility.text)
        facilities.append(f)
    except NoSuchElementException:
        facilities.append('-')       
    
    
    
    
```
